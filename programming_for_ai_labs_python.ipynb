````python
from string import digits
def micadena(num):
  if num == 0: #Special case
    return "0"
  digits = "0123456789" #Conversion table

  result = "" #Initialize string

  negative = num < 0 #Boolean
  num = abs(num)

  while num > 0:
    digit = num % 10 #123 % 10 = 3
    result = digits[digit] + result

    num //=10
  if negative:
    result = "-" + result
  return result


````

````python
def greatest_digits(iterable):

  def suma_digitos(n): #Subfunction
    total = 0 #here we are going to acummulate sum of digits

    while n > 0:
      total += n % 10 #get last digit and sum in total

      n //= 10 #Drop last digit
    return total

  max_num = None

  max_suma = -1

  for numero in iterable:
    suma = suma_digitos(numero)
    if suma > max_suma:
      max_suma = suma
      max_num = numero
  return max_num
````

````python
def real_range(start, stop, step =1.0):
  current = start

  while current <= stop:

    yield current

    current += step
````

````python
for x in real_range(0, 1, 0.1):
  print(x)
````

````python
def get_min(func, a, b, step = 0.001):
  x = a
  min_x = a
  min_value = func(a)

  while x <=b:
    value = func(x)
    if value < min_value:
      min_value = value
      min_x = x
    x += step

  return min_x, min_value
````

````python
def f(x):
  return (x)**2 + 1
min_x, min_value = get_min(f, 0, 5, 0.1)

print(min_x, min_value)
````

````python
def my_range(star, stop = None, step = 1):
  if stop == None:
    stop = star
    star = 0

  if step == 0:
    raise ValueError("step no puede ser cero")

  if step > 0:
    current = star
    while current < stop:
      yield current
      current += step
  else:
    current = star
    while current > stop:
      yield current
      current += step
````

````python
lista = list(range(10,1,-1))
print(lista)
````

````python
lista = list(my_range(10, 1, -1))
print(lista)
````

````python
def next_prime(previous_list):
  num = previous_list[-1] + 1 #next number of last prime number

  while True:
    es_primo = True

    for p in previous_list:
      if num % p == 0:
        es_primo = False
        break

    if es_primo:
      return num
    num +=1


````

````python
it = [2,3,5,6,7]
next_prime(it)
````

````python
def primes():
  num = 2

  while 2:
    es_primo = True
    for i in range(2, num):
      if num % i == 0:
        es_primo = False
        break
    if es_primo:
      yield num
    num += 1
````

````python
for x in primes():
  print(x)
````

````python
def most_frequent_chars(filename, encoding="utf8", n=10):
  with open("prueba.txt", "r", encoding="utf8") as f:
    text = f.read()

  freq = {}
  for char in text:
    if char in freq:
      freq[char] += 1
    else:
      freq[char] = 1
  sorted_freq = sorted(freq.items(), key=lambda x: x[1], reverse=True)
  return sorted_freq[:n]

````

````python
import random
from collections import Counter

def test_randint(a=1, b=6, N=100000):
    results = [random.randint(a, b) for _ in range(N)]
    freq = Counter(results)

    for k in sorted(freq):
        print(f"{k}: {freq[k]}")

test_randint()
````

````python
def test_random(N=100000, bins=10):
    counts = [0] * bins

    for _ in range(N):
        x = random.random()  # entre 0 y 1
        index = int(x * bins)
        if index == bins:  # caso borde
            index -= 1
        counts[index] += 1

    for i, c in enumerate(counts):
        print(f"Bin {i}: {c}")

test_random()
````

````python
class Vector:

    # guardar los valores
    def __init__(self, *components):
        self.data = list(components)

    # cómo se imprime
    def __str__(self):
        return str(tuple(self.data))

    # acceder: v[i]
    def __getitem__(self, i):
        return self.data[i]

    # modificar: v[i] = valor
    def __setitem__(self, i, value):
        self.data[i] = value

    # suma
    def __add__(self, other):
        # vector + vector
        if isinstance(other, Vector):
            if len(self.data) != len(other.data):
                raise ValueError("Dimensiones distintas")

            result = []
            for i in range(len(self.data)):
                result.append(self.data[i] + other.data[i])

            return Vector(*result)

        # vector + número
        else:
            result = []
            for x in self.data:
                result.append(x + other)

            return Vector(*result)

    # resta
    def __sub__(self, other):
        if isinstance(other, Vector):
            if len(self.data) != len(other.data):
                raise ValueError("Dimensiones distintas")

            result = []
            for i in range(len(self.data)):
                result.append(self.data[i] - other.data[i])

            return Vector(*result)
        else:
            result = []
            for x in self.data:
                result.append(x - other)

            return Vector(*result)

    # multiplicación
    def __mul__(self, other):
        if isinstance(other, Vector):
            if len(self.data) != len(other.data):
                raise ValueError("Dimensiones distintas")

            result = []
            for i in range(len(self.data)):
                result.append(self.data[i] * other.data[i])

            return Vector(*result)
        else:
            result = []
            for x in self.data:
                result.append(x * other)

            return Vector(*result)

    # producto escalar
    def dot(self, other):
        if len(self.data) != len(other.data):
            raise ValueError("Dimensiones distintas")

        total = 0
        for i in range(len(self.data)):
            total += self.data[i] * other.data[i]

        return total

    # módulo (longitud)
    def __abs__(self):
        total = 0
        for x in self.data:
            total += x**2

        return total ** 0.5
````

````python
v1 = Vector(1, 2, 3)
v2 = Vector(4, 5, 6)

print(v1)          # (1, 2, 3)
print(v1[0])       # 1

v1[0] = 10
print(v1)          # (10, 2, 3)

print(v1 + v2)     # suma
print(v1 - v2)     # resta
print(v1 * v2)     # multiplicación elemento a elemento

print(v1 + 5)      # vector + número
print(v1 * 2)      # vector * número

print(v1.dot(v2))  # producto escalar

print(abs(v1))     # módulo
````
